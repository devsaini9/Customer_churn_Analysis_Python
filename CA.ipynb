```python
import pandas as pa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pa.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#Replacing blanks with 0 as tenure is 0 and no total cahrges are reocrded

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ",0)
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
        
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv);
```

#Converted 0 & 1 value of senior citizen to make it easier to understand

```python
ax = sns.countplot(x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title("Customer of customer by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn': "count"})
plt.pie(gb['Churn'],labels=gb.index, autopct ="%1.2f%%")
plt.title("Percentage of hurned ustomers")
plt.show()
```

```python
#from the given pie chart we can calculate that26.54% of customers have curned out.
#let"s explore the reason behnd it.
```

```python
plt.figure(figsize=(6,6))
ax = sns.countplot(x='SeniorCitizen', data=df, hue='Churn')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn By SeniorCitizen")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x= 'SeniorCitizen' , data = df, hue = "Churn")
plt.title("Churn By SeniorCitizer")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x= "tenure", data =df , bins=72, hue= "Churn")
plt.show()
```

#people who used our serives for long times have stayed and people who have used our services for 1-2 month have churned

```python
plt.figure(figsize=(3,3))
ax = sns.countplot(x= 'Contract' , data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

```python
#people who have month on month contract are likely to churn then people who have 1-2 year of contract
```

```python
plt.figure(figsize=(9,4))
ax = sns.countplot(x= 'PaymentMethod' , data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customer By Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Customer is likely to churn when he is using electric check as a payment method
```
